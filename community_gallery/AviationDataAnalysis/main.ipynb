```python
from preswald import text, plotly, connect, get_df, table
import pandas as pd
import plotly.express as px
from geopy.geocoders import Nominatim
from collections import defaultdict
import time
from tqdm import tqdm
```

```python
df = pd.read_csv("./data/sample.csv")
```

```python
def get_coordinates(city):
    geolocator = Nominatim(user_agent="city_locator")
    location = geolocator.geocode(city)
    if location:
        return location.latitude, location.longitude
    else:
        return None, None
    
df['city_cleaned'] = df['city'].apply(lambda x: x.split(',')[0])

df['lat'], df['lon'] = zip(*df['city_cleaned'].apply(get_coordinates))
```

```python
ds = defaultdict(tuple)
```

```python
for city in tqdm(df["city_cleaned"].unique()[74:75]): # 68, 69, 74
    print(city)
    ds[city] = get_coordinates(city)
    # time.sleep(.1)
```
